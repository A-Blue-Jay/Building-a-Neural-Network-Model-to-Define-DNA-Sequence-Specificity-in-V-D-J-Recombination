## • H4S2 Sequence Logo
a representation of SARP-seq N(H4-S2) experimental replicates iSeq1, iSeq2, and miSeq1

```python
# Manual frequency weight distribution
frequency_table = np.array([
    [0, 0, 0, 1],  # Position 1 H1
    [1, 0, 0, 0],  # Position 2 H2
    [0, 0, 0, 1],  # Position 3 H3
    [0.25, 0.25, 0.25, 0.25],  # Position 4 H4
    [0.25, 0.25, 0.25, 0.25],  # Position 5 H5
    [0.25, 0.25, 0.25, 0.25],  # Position 6 H6
    [0.25, 0.25, 0.25, 0.25],  # Position 7 H7
    [0.25, 0.25, 0.25, 0.25],  # Position 8 S1
    [0.25, 0.25, 0.25, 0.25],  # Position 9 S2
    [1, 0, 0, 0],  # Position 10 S3
    [0, 0, 0, 1],  # Position 11 S4
    [1, 0, 0, 0],  # Position 12 S5
    [0, 0, 1, 0],  # Position 13 S6
    [1, 0, 0, 0],  # Position 14 S7
    [0, 0, 0, 1],  # Position 15 S8
    [0, 1, 0, 0],  # Position 16 S9
    [0, 0, 1, 0],  # Position 17 S10
    [0, 0, 1, 0],  # Position 18 S11
    [1, 0, 0, 0],  # Position 19 S12
    [1, 0, 0, 0],  # Position 20 N1
    [0, 0, 0, 1],  # Position 21 N2
    [1, 0, 0, 0],  # Position 22 N3
    [1, 0, 0, 0],  # Position 23 N4
    [1, 0, 0, 0],  # Position 24 N5
    [1, 0, 0, 0],  # Position 25 N6
    [1, 0, 0, 0],  # Position 26 N7
    [0, 0, 0, 1],  # Position 27 N8
    [0, 0, 0, 1]  # Position 28 N9
])

# Create a DataFrame from the frequency table
df = pd.DataFrame(frequency_table, columns=['A', 'T', 'G', 'C'])

# Custom color scheme dictionary
custom_colors = {'A': 'green', 'T': 'red', 'G': 'black', 'C': 'blue'}

# Create the logo
logo = logomaker.Logo(df, color_scheme = custom_colors, figsize =([10, 2]))

# Style the logo
logo.ax.set_title("Sequence Logo")
logo.style_spines(visible=True)  # Hide spines
logo.highlight_position_range(pmin=3, pmax=8, color='gold')

plt.xlabel('Position')
plt.ylabel('Probability')
plt.title('SARP-seq H4S2 input library')
# Add x tick labels
x_ticks = ['H1', 'H2', 'H3', 'H4', 'H5', 'H6', 'H7', 'S1', 'S2', 'S3', 'S4', 'S5', 'S6', 'S7', 'S8', 'S9', 'S10', 'S11', 'S12', 'N1', 'N2', 'N3', 'N4', 'N5', 'N6', 'N7', 'N8', 'N9']
plt.xticks(range(len(df)), x_ticks, rotation=90, fontsize=12)

# Show the logo
plt.show()
```
